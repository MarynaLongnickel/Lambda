```

import numpy as np
import  matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm


def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
print(X.shape, '\n', Y.shape, '\n', Z.shape, '\n')

ax = plt.axes(projection = '3d')
ax.plot_surface(X, Y, Z, rstride = 1, cstride = 1, cmap = 'viridis', edgecolor = 'none');

# for angle in range(0, 360):
#     ax.view_init(30, angle)
#     plt.show()

# ax = plt.axes(projection='3d')
# ax.plot_trisurf(X, Y, Z, cmap='viridis', edgecolor='none');


# The linear spaced vectors x and y defined by linspace consist of 50 and 40 points respectively.
# These vectors are then used by the meshgrid function to create a 2D linear spaced point cloud.
# This will be a 40 x 50 grid of points for each of the x and y coordinates.
# The output of the function meshgrid creates an indexing matrix that holds in each cell the x and y coordinates for each point of space.
```

```
# More 3d plots!
from numpy import pi, sin, cos, mgrid, sqrt

x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)

X, Y = np.meshgrid(x, y)
R = sqrt(X**2 + Y**2)
Z = sin(R)
W = sin(X+Y)**2 + X**2 - Y**2
T = W + cos(X**2 - Y**2)

ax = plt.axes(projection = '3d')
ax.plot_surface(X, Y, Z, rstride = 1, cstride = 1, cmap = 'inferno', edgecolor = 'none')
plt.show()

ax = plt.axes(projection = '3d')
ax.plot_surface(X, Y, T, rstride = 1, cstride = 1, cmap = 'viridis', edgecolor = 'none')
plt.show()

# ---------------------------------------------------------------------------------------------

dphi, dtheta = pi/25.0, pi/25.0
[phi,theta] = mgrid[0:pi+dphi*1.5:dphi,0:2*pi+dtheta*1.5:dtheta]
m0 = 4; m1 = 3; m2 = 2; m3 = 3; m4 = 6; m5 = 2; m6 = 6; m7 = 4;
r = sin(m0*phi)**m1 + cos(m2*phi)**m3 + sin(m4*theta)**m5 + cos(m6*theta)**m7
A = r*sin(phi)*cos(theta)
B = r*cos(phi)
C = r*sin(phi)*sin(theta)

ax = plt.axes(projection = '3d')
ax.plot_surface(A, B, C, rstride = 1, cstride = 1, cmap = 'magma', edgecolor = 'none')
plt.show()

# ---------------------------------------------------------------------------------------------

def surf(u, v):
    half = (0 <= u) & (u < pi)
    r = 4*(1 - cos(u)/2)
    x = 6*cos(u)*(1 + sin(u)) + r*cos(v + pi)
    x[half] = (
        (6*cos(u)*(1 + sin(u)) + r*cos(u)*cos(v))[half])
    y = 16 * sin(u)
    y[half] = (16*sin(u) + r*sin(u)*cos(v))[half]
    z = r * sin(v)
    return x, y, z

u, v = np.linspace(0, 2*pi, 40), np.linspace(0, 2*pi, 40)
ux, vx =  np.meshgrid(u,v)
x, y, z = surf(ux, vx)

fig = plt.figure()
ax = fig.gca(projection = '3d')
plot = ax.plot_surface(x, y, z, rstride = 1, cstride = 1, cmap = 'ocean',
                       linewidth = 0, antialiased = False)

plt.show()
```
